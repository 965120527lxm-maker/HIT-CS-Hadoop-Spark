> 问题：对于一个数据流，统计每个元素出现的频率。但是，数据流很大，我的内存可能都存不下一个表。

所以，我们需要节省每一个数计数器所用的存储空间

Morris(1978)：我可以通过做实验的思想，用一个小整数C来近似地表达计数值；

Flajolet-Martin(1985)：我继承Morris的想法，“实验次数越多，结果越小众”。我能给一个随机hash函数，把数映射到一个64位（或者其它）的二进制数，然后看看这个二进制数末尾连续0的个数；显然，N越大，我越可能出现长的个数。

Loglog(2004)，HyperLoglog(2007):我们总结了算法。利用调和平均值；把hash空间分成m个桶；每个桶独立运行；综合桶的结果。

减少了算法的方差。
